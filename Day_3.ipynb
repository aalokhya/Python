```python

# 1. rev the array

# sol 1
arr = [1,2,3,4,5]
left = 0
right = len(arr) - 1 # as arr index start from 0 (5 - 1 = 4)
while left < right:
    arr[left], arr[right] = arr[right], arr[left]
    left += 1
    right -= 1
print(arr)


# sol 2
arr = list(map(int,input().split()))
a=arr[::-1]
print(*a)
```

```python
# 2. find max and min ele in arr

# sol 1

arr = [9,8,5,1,2]
max = arr[0]
min = arr[0]
for i in range(len(arr)):
  if arr[i] > max:
    max = arr[i]
  if arr[i] < min:
    min = arr[i]
print("min = ", min, "max = ",max)

# sol 2

arr = [9, 8, 5, 1, 2]

sorted_arr = sorted(arr) # use sorted fun the return frst and last ele
minimum = sorted_arr[0]
maximum = sorted_arr[-1]

print("Max:", maximum)
print("Min:", minimum)

# sol 3

arr = [9, 8, 5, 1, 2]

maximum = max(arr)   # max and min fun of python
minimum = min(arr)

print("Maximum:", maximum)
print("Minimum:", minimum)

# sol 4

import numpy as np

arr = np.array([9, 8, 5, 1, 2])

maximum = np.max(arr)
minimum = np.min(arr)

print("Maximum:", maximum)
print("Minimum:", minimum)



```

```python
# 4. count even and odd no. and give cout

arr = [2,9,7,5,4,3,1,0]
c_e = 0
c_o = 0
for i in range(len(arr)):
  if arr[i] % 2 == 0:
    print(arr[i], "is even")
    c_e += 1
  else:
    print("odd")
    c_o +=1
print("even = ", c_e, "odd = ", c_o)

# sol 2

arr = [9, 8, 5, 1, 2, 6]

even_count = len([num for num in arr if num % 2 == 0])
odd_count = len([num for num in arr if num % 2 != 0])

print("Even numbers:", even_count)
print("Odd numbers:", odd_count)

```

```python
# 6. find freq of each ele or given ele in arr

arr = [6,8,2,2,4,5]
freq = {}
for num in arr:
  if num in freq:
        freq[num] += 1
  else:
        freq[num] = 1

for key, value in freq.items():
    print(f"{key} → {value} times")

# ask for specific element

arr = [2, 3, 2, 4, 3, 2]
value = 2

count = 0
for num in arr:
    if num == value:
        count += 1

print(f"Frequency of {value} is {count}")

# sol 2

# for all ele in arr

from collections import Counter
arr = [2, 3, 2, 4, 3, 2]
freq = Counter(arr)
print(freq)


# for specific

arr = [2, 3, 2, 4, 3, 2]
value = 2

count = arr.count(value)

print(f"Frequency of {value} is {count}")


```

```python
# 7. find freq of each ele and retrun where freq == ele

arr = [2,2,4,5,5,7,1,9,9]
freq = {}
for i in arr:
  if i in freq:
    freq[i] += 1
  else:
    freq[i] = 1

for key,value in freq.items():
  if value == key:
    print(key)

```
