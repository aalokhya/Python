```python
# find max freq from given arr

arr = list(map(int,input().split()))
freq = {}
for num in arr:
  if num in freq:
    freq[num] += 1
  else:
    freq[num] = 1
print(freq)

max_freq = 0
ele = None
for key,value in freq.items():
  if value > max_freq:
    max_freq = value
    ele = key
print(" number : ", ele, ' '," max_freq :" ,max_freq)
```

```python
# take an arr do sq of each no. then add first num then subtract apply that pattern for overall arr

arr = list(map(int, input().split()))

sq = []
for num in arr:
    sq.append(num ** 2)

result = 0
for i in range(len(arr)):
    if i % 2 == 0:
        result += sq[i]
    else:
        result -= sq[i]

print(result)
```
