```python
'''

# 1. move all zeros to end of the list

# Sol- 1: create 2 more lists names arr1 and 2 then by seeing or iterating through the each element in the original input arr if element is 0 add it to arr1 if not
# append to arr2 finally join arr2 to arr1 and return arr

'''

arr = list(map(int,input().split()))

# In the above statement we use input() to take values as str (1 2 3) then split() i.e, ["1","2","3"] then map function applies int() to all the values so that they get converted
# into int values map(int, ['1', '2','3']) then list() to make them as complete list i.e, arr = [1, 2, 3]

arr1 = []
arr2 = []

# when we use for i in arr here i is the element in the arr but when it comes to for i in range(len(arr)) i is the index of the elements in arr so to access that we use
# arr[i] but in this case as it is for i in arr which is direct element we can use i and do the work or continue in below steps

for i in arr:
  if i == 0:
    arr1.append(i)
  else:
    arr2.append(i)
arr = arr2 + arr1

print(arr) # prints list
print(*arr) # unpacking operator which prints as nrml no.


 # Time Complexity : O(n) as there is an loop
 # space : O(n) as we craeted 2 seperate arr again and used it

'''
 # Sol - 2 : "TWO POINTER" so t decrease the complexity we acn do it without creating another arr by reading the question they say move zero elements to end its same as moving all non-zero elements
 # to left so we will take a left,right pointers if we see a non-zero value swap both and then move left pointer. here only when we see a non zero element left pointer moves but
 # right moves everytime for each step iterates troughout arr and which we dont need to specify seperately as in for loop its done.

'''

arr = [0,1,0,3,12]
l = 0
for r in range(len(arr)):
  # in previous way we used the iteration on values directly i.e, i but now we are using the index values moving the pointers so we use this loop
  # with the len range and also accessing elements using arr[index]
  if arr[r]!= 0:
    arr[l],arr[r] = arr[r],arr[l]
    l+= 1

print(arr)

# Time : O(n) as there is a loop
# space : O(1) as we did not create any extra list


```

```python
'''
2. Given an array nums of size n, return the majority element.The majority element is the element that appears more than ⌊n / 2⌋ times.
   You may assume that the majority element always exists in the array.

Sol - 1 : they are saying to find the most repeated num or the num with highest freq in the arr s for this we will first find freq of each no. that is count of each no. by
creating a dict freq = {} which stores key,values that is element,count of each number then we will find out the max value from that dict that is
which key or number has the high value return that

'''

arr = list(map(int,input().split()))
freq = {}
for num in arr:
  if num in freq:
    freq[num] += 1
  else:
    freq[num] = 1
print(freq) # returns all the numbers with their count

max_freq = 0
ele = None
for key,value in freq.items():
  # used to iterate in the dict items
  if value > max_freq:
    max_freq = value
    ele = key
print(ele)

# time : O(n) as We iterate to build the frequency dictionary and to find the maximum
# space : O(n) In the worst case, we store each element in the dictionary if all elements are unique

''' Sol - 2 : using the predefined python library counter  '''

from collections import Counter

arr = list(map(int, input().split())) # arr = [3, 3, 4, 3, 2]

freq = Counter(arr) # it changes the freq to counter obj that is a dict , Counter ({3: 3, 4: 1, 2: 1})
# freq.most_common(1) # returns a list of tuple [(element, frequency) which is most common here it is [(3,3)]

print(freq.most_common(1)[0][0])

# [0] → selects the first tuple (3, 3)
# [0] again → selects the element from that tuple → 3

'''
freq.most_common(1)   → [(3, 3)]       # list of tuple(s)
freq.most_common(1)[0] → (3, 3)        # first tuple
freq.most_common(1)[0][0] → 3          # element
freq.most_common(1)[0][1] → 3          # count
'''

# same time,space complexity that is O(n)

'''
sol - 3: This is " Boyer-Moore Voting Algorithm " where we will start by taking an var candidate that is each ele in arr and a count var if the same element repeats we will
increase the count by 1 if different num appears we will decrease it by 1 and at last print the count to get majority freq and candidate to get that element
'''
arr = list(map(int, input().split()))

candidate = None
count = 0

for num in arr:
    if count == 0:
        candidate = num
        count = 1
    elif num == candidate:
        count += 1
    else:
        count -= 1

freq = arr.count(candidate) # count how many times it appeared

print("Maj ele : " , candidate , "freq : ",freq)

# time : O(n) as We traverse the array twice, to find the candidate to count its occurrences O(n + n) = O(n).
# Space Complexity O(1): We only use var and no xtra arr or dict
```
