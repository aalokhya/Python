```python
'''
1.Given an integer array nums sorted in non-decreasing order, remove the duplicates in-place such that each unique element appears only once. The relative order of the elements should be kept the same. Then return the number of unique elements in nums.
Consider the number of unique elements of nums to be k, to get accepted, you need to do the following things:
Change the array nums such that the first k elements of nums contain the unique elements in the order they were present in nums initially. The remaining elements of nums are not important as well as the size of nums.
Return k.
'''
# in place means we should not use any other extra space means other arr or lists in solution

'''
# sol-1 : 2 pointer approach is best method for this where we take i = 0 and j = 1 till len of arr then compare i and j if both are same move j if not same move i then copy value of j to i
# this helps to get the arr with unique elements in place '''

arr = list(map(int, input().split()))
i = 0
for j in range(1, len(arr)):
    if arr[i] != arr[j]:
        i += 1
        arr[i] = arr[j]       # Move it next to previous unique element

# Replace remaining positions with "_"
for k in range(i + 1, len(arr)):
    arr[k] = "_"
print( i + 1)
print( arr)

# time complexity O(n)
# space O(1)

 ''' sol-2 : use set (not suitable for inplace prblms) '''

 def removeDuplicates(nums):
    # Convert to set to remove duplicates
    unique = sorted(set(nums))  # set removes dups, sorted keeps order
    nums[:len(unique)] = unique  # put unique values back into original list
    return len(unique)

    # time : O(log(n))
    # space : O(n)

''' sol-3 :  create new list '''

def removeDuplicates(nums):
    unique = []  # new list to hold unique elements
    for num in nums:
        if not unique or unique[-1] != num:  # check with last added
            unique.append(num)
    nums[:len(unique)] = unique  # copy back to original
    return len(unique)

    # time & space : O(n)





```
